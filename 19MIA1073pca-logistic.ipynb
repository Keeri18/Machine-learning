```python
X = data2.iloc[:,0:65]
y = data2.iloc[:,-1]
```

```python
X, X_test, y, y_test = train_test_split(X, y, train_size = 0.795770729, random_state = 1)
```

```python
X1 = sc.fit_transform(X)
X_test1 = sc.transform(X_test)
```

```python
lr = LogisticRegression(max_iter = 1000)
```

```python
lr.fit(X1,y)
```

```python
y_pred = lr.predict(X_test1)
```

```python
score = accuracy_score(y_test, y_pred)
```
